```python
import ast

def safe_list_to_text(x):
    if pd.isna(x):
        return ""
    try:
        return " ".join(ast.literal_eval(x))
    except:
        return str(x)

products_clean["ingredients"] = products_clean["ingredients"].apply(safe_list_to_text)
products_clean["highlights"] = products_clean["highlights"].apply(safe_list_to_text)

products_clean[["ingredients", "highlights"]].head(3)

# Amaç:
# ingredients ve highlights sütunlarında liste gibi görünen string verileri
# düz metne dönüştürmek.
# Boş (NaN) değerler için boş string döndürülür.
# Böylece bu alanlar daha sonra product_text oluştururken temiz ve kullanılabilir hale gelir.
```

```python
text_cols = [
    "product_name",
    "brand_name",
    "primary_category",
    "secondary_category",
    "tertiary_category",
    "highlights",
    "ingredients"
]

products_clean["product_text"] = (
    products_clean[text_cols]
    .fillna("")
    .agg(" ".join, axis=1)
)

products_clean[["product_name", "product_text"]].head(3)

# Amaç:
# Ürünü temsil edecek tek bir metin alanı (product_text) oluşturmak.
# product_name, brand_name, kategori bilgileri, highlights ve ingredients
# sütunları birleştirilerek her ürün için açıklayıcı bir metin oluşturulur.
# Eksik değerler boş string ile doldurulur ve tüm alanlar tek metin halinde birleştirilir.
```

```python
reviews_profile = reviews_clean[
    ["product_id", "rating", "skin_type", "skin_tone", 
     "eye_color", "hair_color", "is_recommended"]
].copy()

print("Shape:", reviews_profile.shape)
reviews_profile.head()

# Amaç:
# Profil tabanlı öneri sistemi için gerekli kullanıcı özelliklerini içeren
# review kolonlarını ayırarak yeni bir veri seti (reviews_profile) oluşturmak.
# Bu tablo ürün puanı ile birlikte kullanıcı cilt tipi, ten tonu, göz ve saç rengi gibi
# özellikleri içerir ve ileride ürün-profili skorlarının hesaplanmasında kullanılacaktır.
```

```python
# Skin type boş olanları ayrı değerlendirebiliriz, ama şimdilik dahil etmiyoruz
reviews_profile_nonnull = reviews_profile.dropna(subset=["skin_type"])

product_skin_stats = (
    reviews_profile_nonnull
    .groupby(["product_id", "skin_type"])
    .agg(
        mean_rating=("rating", "mean"),
        review_count=("rating", "count"),
        recommend_rate=("is_recommended", "mean")
    )
    .reset_index()
)

product_skin_stats.head()

# Amaç:
# Skin type bilgisi bulunan review'ları kullanarak ürünlerin farklı cilt tipleri
# için nasıl performans gösterdiğini hesaplamak.
# Her product_id ve skin_type kombinasyonu için:
# - ortalama puan (mean_rating)
# - review sayısı (review_count)
# - önerilme oranı (recommend_rate)
# hesaplanarak product_skin_stats tablosu oluşturulur.
```

```python
product_skin_pivot = (
    product_skin_stats
    .pivot(index="product_id", columns="skin_type", values="mean_rating")
    .reset_index()
)

product_skin_pivot.head()

# Amaç:
# product_skin_stats tablosunu pivot ederek her ürün için cilt tiplerine göre
# ortalama puanların ayrı sütunlarda yer aldığı bir tablo oluşturmak.
# Böylece her product_id için combination, dry, normal ve oily cilt tiplerine
# ait ortalama rating değerleri kolayca erişilebilir hale gelir.
```

```python
# Global ortalama rating
global_mean = reviews_profile_nonnull["rating"].mean()

# m değeri (güven sabiti)
m = 20  # başlangıç için makul bir değer

product_skin_weighted = (
    reviews_profile_nonnull
    .groupby(["product_id", "skin_type"])
    .agg(
        mean_rating=("rating", "mean"),
        review_count=("rating", "count")
    )
    .reset_index()
)

# Düzeltilmiş skor hesapla
product_skin_weighted["weighted_score"] = (
    (product_skin_weighted["mean_rating"] * product_skin_weighted["review_count"] +
     global_mean * m)
    /
    (product_skin_weighted["review_count"] + m)
)

product_skin_weighted.head()

# Amaç:
# Ürünlerin cilt tipi bazındaki ortalama puanlarını, review sayısını da dikkate alarak
# daha güvenilir bir weighted_score ile düzeltmek.
# Bu yaklaşım, az sayıda review alan ürünlerin ortalama puanlarının yanıltıcı şekilde
# çok yüksek veya çok düşük görünmesini engeller.
```

```python
product_skin_weighted_pivot = (
    product_skin_weighted
    .pivot(index="product_id", columns="skin_type", values="weighted_score")
    .reset_index()
)

product_skin_weighted_pivot.head()

# Amaç:
# product_skin_weighted tablosunu pivot ederek her ürün için cilt tiplerine göre
# weighted_score değerlerini ayrı sütunlarda göstermek.
# Böylece her product_id için dry, oily, normal ve combination cilt tiplerine ait
# güvenilir skorlar tek satırda tutulur.
```

```python
product_skin_count_pivot = (
    product_skin_weighted
    .pivot(index="product_id", columns="skin_type", values="review_count")
    .reset_index()
)

product_skin_count_pivot.head()

# Amaç:
# product_skin_weighted tablosundaki review_count değerlerini pivot ederek
# her ürün için cilt tiplerine göre kaç review bulunduğunu ayrı sütunlarda göstermek.
# Böylece combination, dry, normal ve oily cilt tipleri için review sayıları
# tek satırda ürün bazında tutulur.
```

```python
product_skin_profile = product_skin_weighted.merge(
    product_skin_count_pivot,
    on="product_id",
    how="left",
    suffixes=("", "_count")
)

product_skin_profile.head()

# Amaç:
# Skin type bazlı weighted skor tablosunu, yine skin type bazlı review sayılarıyla birleştirmek.
# Böylece her ürün-cilt tipi satırında hem weighted_score bilgisi
# hem de ürünün tüm cilt tiplerindeki review sayıları birlikte görülebilir.
```

```python
product_skin_final = product_skin_weighted_pivot.merge(
    product_skin_count_pivot,
    on="product_id",
    how="left",
    suffixes=("_score", "_count")
)

product_skin_final.head()

# Skin type bazlı ürün skorları (product_skin_weighted_pivot) ile
# skin type bazlı review sayıları (product_skin_count_pivot) product_id üzerinden birleştirilir.
# Bu tablo profile-based recommendation aşamasında kullanılacak ürün-cilt tipi profilini temsil eder.
```

```python
# skin_tone boş olanları bu analizde kullanmayacağız
reviews_tone = reviews_clean[["product_id", "rating", "skin_tone", "is_recommended"]].copy()
reviews_tone = reviews_tone.dropna(subset=["skin_tone"])

global_mean_tone = reviews_tone["rating"].mean()
m = 20  # temkinli sistem sabiti (aynı)

product_tone_weighted = (
    reviews_tone
    .groupby(["product_id", "skin_tone"])
    .agg(
        mean_rating=("rating", "mean"),
        review_count=("rating", "count"),
        recommend_rate=("is_recommended", "mean")
    )
    .reset_index()
)

product_tone_weighted["weighted_score"] = (
    (product_tone_weighted["mean_rating"] * product_tone_weighted["review_count"] +
     global_mean_tone * m)
    /
    (product_tone_weighted["review_count"] + m)
)

product_tone_weighted.head()

# Amaç:
# Skin tone bilgisi bulunan review’ları kullanarak ürünlerin farklı ten tonlarında
# nasıl performans gösterdiğini hesaplamak.
# Her product_id ve skin_tone kombinasyonu için:
# - ortalama puan (mean_rating)
# - review sayısı (review_count)
# - önerilme oranı (recommend_rate)
# hesaplanır ve daha sonra güvenilirliği artırmak için weighted_score üretilir.
```

```python
# 1) Tone skor pivot
product_tone_score_pivot = (
    product_tone_weighted
    .pivot(index="product_id", columns="skin_tone", values="weighted_score")
    .reset_index()
)

# 2) Tone count pivot
product_tone_count_pivot = (
    product_tone_weighted
    .pivot(index="product_id", columns="skin_tone", values="review_count")
    .reset_index()
)

# Kolon isimlerini ayırt edelim
product_tone_score_pivot = product_tone_score_pivot.rename(
    columns={c: f"{c}_tone_score" for c in product_tone_score_pivot.columns if c != "product_id"}
)
product_tone_count_pivot = product_tone_count_pivot.rename(
    columns={c: f"{c}_tone_count" for c in product_tone_count_pivot.columns if c != "product_id"}
)

# 3) Hepsini product_id üzerinden birleştir
product_profile = (
    product_skin_final
    .merge(product_tone_score_pivot, on="product_id", how="left")
    .merge(product_tone_count_pivot, on="product_id", how="left")
)

print("product_profile shape:", product_profile.shape)
product_profile.head()

# Amaç:
# Skin tone bazlı weighted skorları ve review sayılarını pivot ederek ürün seviyesine taşımak
# ve bunları skin type bazlı ürün profil tablosu ile birleştirerek nihai product_profile
# tablosunu oluşturmak.
# Bu tablo, her ürün için farklı kullanıcı özelliklerine (skin_type ve skin_tone)
# göre performans skorlarını ve veri yoğunluğunu içerir.
```

```python
# ------------------------------------------------------------------
# PROFILE-BASED + FALLBACK RECOMMENDATION FONKSİYONU
#
# Bu fonksiyon, kullanıcının profil bilgilerine göre ürün önerir.
#
# Ana mantık:
# 1) Önce ürünleri product_profile tablosundan kategoriye göre filtreler
# 2) Eğer bu kategoride profile-based skor üretilebilen ürün varsa:
#       - skin_type ve skin_tone skorlarını kullanır
#       - dinamik ağırlıklarla final profile_score hesaplar
# 3) Eğer profile-based ürün yoksa:
#       - products_clean içinden aynı kategorideki ürünleri alır
#       - rating, reviews ve loves_count ile fallback skor üretir
#
# Böylece sistem bazı kategorilerde profile verisi olmasa bile
# öneri üretebilir.
# ------------------------------------------------------------------

def recommend_products(
    skin_type,
    skin_tone,
    category,
    top_n=10
):
    import numpy as np

    # ---------------------------------------------------------------
    # 1) Ürün profil tablosu ile temiz ürün tablosunu birleştir
    # ---------------------------------------------------------------
    df = product_profile.merge(
        products_clean,
        on="product_id",
        how="left"
    )

    # ---------------------------------------------------------------
    # 2) Kategoriye göre filtrele
    # contains kullanıyoruz çünkü kategori adı daha uzun olabilir
    # Örn: "Under-Eye Concealer" içinde "concealer" geçer
    # ---------------------------------------------------------------
    df = df[
        df["tertiary_category"]
        .fillna("")
        .str.lower()
        .str.contains(category.lower(), na=False)
    ]

    # ---------------------------------------------------------------
    # 3) Eğer product_profile içinde bu kategori yoksa
    # fallback recommendation çalıştır
    # ---------------------------------------------------------------
    if df.shape[0] == 0:
        print("Profile-based veri bulunamadı. Fallback recommendation çalışıyor...")

        fallback_df = products_clean[
            products_clean["tertiary_category"]
            .fillna("")
            .str.lower()
            .str.contains(category.lower(), na=False)
        ].copy()

        if fallback_df.shape[0] == 0:
            print("Bu kategori için ürün bulunamadı.")
            return None

        # -----------------------------------------------------------
        # Fallback skor:
        # rating, reviews ve loves_count bilgilerini kullanıyoruz
        # Eksik değerleri güvenli şekilde dolduruyoruz
        # -----------------------------------------------------------
        fallback_df["rating"] = fallback_df["rating"].fillna(0)
        fallback_df["reviews"] = fallback_df["reviews"].fillna(0)
        fallback_df["loves_count"] = fallback_df["loves_count"].fillna(0)

        # log1p: büyük loves_count değerlerini yumuşatmak için
        fallback_df["fallback_score"] = (
            0.5 * fallback_df["rating"] +
            0.3 * np.log1p(fallback_df["reviews"]) +
            0.2 * np.log1p(fallback_df["loves_count"])
        )

        fallback_df = fallback_df.sort_values("fallback_score", ascending=False)

        result_cols = [
            "product_id",
            "product_name",
            "brand_name",
            "primary_category",
            "secondary_category",
            "tertiary_category",
            "price_usd",
            "rating",
            "reviews",
            "loves_count",
            "fallback_score"
        ]

        result_cols = [col for col in result_cols if col in fallback_df.columns]

        return fallback_df[result_cols].head(top_n)

    # ---------------------------------------------------------------
    # 4) Kullanıcının profiline göre ilgili skor ve count sütunlarını seç
    # Örn:
    # skin_type = dry   -> dry_score, dry_count
    # skin_tone = light -> light_tone_score, light_tone_count
    # ---------------------------------------------------------------
    type_score_col = f"{skin_type.lower()}_score"
    type_count_col = f"{skin_type.lower()}_count"

    tone_score_col = f"{skin_tone}_tone_score"
    tone_count_col = f"{skin_tone}_tone_count"

    if type_score_col not in df.columns:
        print(f"Cilt tipi için uygun skor sütunu bulunamadı: {type_score_col}")
        return None

    if tone_score_col not in df.columns:
        print(f"Ten tonu için uygun skor sütunu bulunamadı: {tone_score_col}")
        return None

    # ---------------------------------------------------------------
    # 5) İlgili skor ve count sütunlarını standart isimlerle kopyala
    # ---------------------------------------------------------------
    df["type_score"] = df[type_score_col]
    df["type_count"] = df[type_count_col]

    df["tone_score"] = df[tone_score_col]
    df["tone_count"] = df[tone_count_col]

    # ---------------------------------------------------------------
    # 6) Dinamik ağırlık hesaplama
    #
    # Mantık:
    # - Veri sayısı yüksek olan profile sinyaline daha çok güven
    # - Ancak count çok büyükse sistemi domine etmesin diye sqrt kullan
    # - Başlangıçta skin_type daha önemli kabul ediliyor
    # ---------------------------------------------------------------
    type_strength = np.sqrt(df["type_count"].fillna(0))
    tone_strength = np.sqrt(df["tone_count"].fillna(0))

    base_type_weight = 0.7
    base_tone_weight = 0.3

    weighted_type_strength = base_type_weight * type_strength
    weighted_tone_strength = base_tone_weight * tone_strength

    total_strength = weighted_type_strength + weighted_tone_strength
    total_strength = total_strength.replace(0, 1)

    df["dynamic_type_weight"] = weighted_type_strength / total_strength
    df["dynamic_tone_weight"] = weighted_tone_strength / total_strength

    # Eğer skor eksikse o özelliğin ağırlığını sıfırla
    df.loc[df["type_score"].isna(), "dynamic_type_weight"] = 0
    df.loc[df["tone_score"].isna(), "dynamic_tone_weight"] = 0

    # Ağırlıkları tekrar normalize et
    weight_sum = df["dynamic_type_weight"] + df["dynamic_tone_weight"]
    weight_sum = weight_sum.replace(0, 1)

    df["dynamic_type_weight"] = df["dynamic_type_weight"] / weight_sum
    df["dynamic_tone_weight"] = df["dynamic_tone_weight"] / weight_sum

    # ---------------------------------------------------------------
    # 7) Final profile score hesapla
    # ---------------------------------------------------------------
    df["profile_score"] = (
        df["dynamic_type_weight"] * df["type_score"].fillna(0) +
        df["dynamic_tone_weight"] * df["tone_score"].fillna(0)
    )

    # ---------------------------------------------------------------
    # 8) Ürünleri final profile_score'a göre sırala
    # ---------------------------------------------------------------
    df = df.sort_values("profile_score", ascending=False)

    # ---------------------------------------------------------------
    # 9) Sonuç tablosunda gösterilecek sütunları seç
    # ---------------------------------------------------------------
    result_cols = [
        "product_id",
        "product_name",
        "brand_name",
        "primary_category",
        "secondary_category",
        "tertiary_category",
        "price_usd",
        "type_score",
        "tone_score",
        "dynamic_type_weight",
        "dynamic_tone_weight",
        "profile_score"
    ]

    result_cols = [col for col in result_cols if col in df.columns]

    return df[result_cols].head(top_n)
```

```python
# ---------------------------------------------------------------
# CONTENT-BASED RECOMMENDATION MODEL
#
# Bu model ürünlerin metin içeriklerini kullanarak
# birbirine benzer ürünleri bulur.
#
# product_text sütunu şu bilgileri içerir:
# - product_name
# - brand_name
# - category bilgileri
# - highlights / ingredients
#
# Bu metinler TF-IDF ile sayısal vektörlere çevrilir.
# Daha sonra cosine similarity ile ürün benzerliği hesaplanır.
# ---------------------------------------------------------------

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# ---------------------------------------------------------------
# TF-IDF VECTORIZATION
#
# product_text sütununu sayısal vektörlere dönüştürüyoruz.
# Böylece her ürün bir feature vektörü haline geliyor.
# ---------------------------------------------------------------

tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=5000
)

tfidf_matrix = tfidf.fit_transform(products_clean["product_text"])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python
# ---------------------------------------------------------------
# COSINE SIMILARITY MATRIX
#
# Bu adımda tüm ürünler arasındaki benzerliği hesaplıyoruz.
# Her ürün diğer tüm ürünlerle karşılaştırılır.
#
# Sonuç:
# similarity_matrix[i][j]
# → ürün i ile ürün j ne kadar benzer
# ---------------------------------------------------------------

similarity_matrix = cosine_similarity(tfidf_matrix)

print("Similarity matrix shape:", similarity_matrix.shape)
```

```python
similarity_matrix = cosine_similarity(tfidf_matrix)
print("Similarity matrix shape:", similarity_matrix.shape)
```

```python
# ---------------------------------------------------------------
# CONTENT-BASED RECOMMENDATION FUNCTION
#
# Bu fonksiyon verilen bir ürüne en benzer ürünleri bulur.
#
# Çıktıya özellikle şu alanları ekliyoruz:
# - product_id
# - similarity_score
#
# Çünkü hybrid recommendation aşamasında content skoru ile
# collaborative filtering skorunu product_id üzerinden birleştireceğiz.
# ---------------------------------------------------------------

def recommend_similar_products(product_name, top_n=10):

    # -----------------------------------------------------------
    # 1) Ürünün dataset içindeki index'ini bul
    # -----------------------------------------------------------
    matches = products_clean[
        products_clean["product_name"].str.lower().str.contains(product_name.lower(), na=False)
    ]

    if matches.shape[0] == 0:
        print("Ürün bulunamadı.")
        return None

    product_index = matches.index[0]

    # -----------------------------------------------------------
    # 2) Bu ürünün similarity skorlarını al
    # -----------------------------------------------------------
    similarity_scores = list(enumerate(similarity_matrix[product_index]))

    # -----------------------------------------------------------
    # 3) En yüksek benzerlik skorlarına göre sırala
    # -----------------------------------------------------------
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # İlk eleman ürünün kendisi olduğu için atlıyoruz
    similarity_scores = similarity_scores[1:top_n+1]

    # -----------------------------------------------------------
    # 4) Benzer ürünlerin indexlerini ve similarity skorlarını al
    # -----------------------------------------------------------
    product_indices = [i[0] for i in similarity_scores]
    sim_scores = [i[1] for i in similarity_scores]

    # -----------------------------------------------------------
    # 5) Sonuç tablosu
    # product_id ve similarity_score özellikle ekleniyor
    # -----------------------------------------------------------
    result = products_clean.iloc[product_indices][
        [
            "product_id",
            "product_name",
            "brand_name",
            "primary_category",
            "secondary_category",
            "tertiary_category",
            "price_usd",
            "rating",
            "reviews",
            "loves_count"
        ]
    ].copy()

    result["similarity_score"] = sim_scores

    return result
```

```python
# ------------------------------------------------------------------
# COLLABORATIVE FILTERING İÇİN ANA VERİ SETİNİ YÜKLEME
#
# Bu adımda daha önce preprocessing aşamasında oluşturulan
# reviews_cf_last.csv dosyasını yüklüyoruz.
#
# Bu tablo:
# - author_id  -> kullanıcı
# - product_id -> ürün
# - rating     -> puan
# - submission_time -> en güncel review zamanı
#
# Aynı kullanıcı aynı ürüne birden fazla review verdiyse,
# bu tabloda sadece en güncel review tutulmuştur.
# ------------------------------------------------------------------

cf_data = pd.read_csv("../data_interim/reviews_cf_last.csv")

print("CF data shape:", cf_data.shape)
cf_data.head()
```

```python
# ------------------------------------------------------------------
# CF VERİ SETİNİN TEMEL YAPISINI İNCELEME
#
# Bu adımda:
# - kaç benzersiz kullanıcı var
# - kaç benzersiz ürün var
# - toplam rating sayısı kaç
# bunları inceliyoruz.
# ------------------------------------------------------------------

n_users = cf_data["author_id"].nunique()
n_items = cf_data["product_id"].nunique()
n_ratings = len(cf_data)

print("Benzersiz kullanıcı sayısı:", n_users)
print("Benzersiz ürün sayısı:", n_items)
print("Toplam rating sayısı:", n_ratings)
```

```python
# ------------------------------------------------------------------
# SPARSITY (SEYREKLİK) HESAPLAMA
#
# User-item matrisi çoğunlukla boş olur.
# Bu yüzden CF veri setinin ne kadar seyrek olduğunu hesaplıyoruz.
#
# sparsity = 1 - (mevcut rating sayısı / tüm olası user-item çiftleri)
# ------------------------------------------------------------------

sparsity = 1 - (n_ratings / (n_users * n_items))

print("Sparsity:", round(sparsity, 4))
print("Sparsity (%):", round(sparsity * 100, 2))
```

```python
# ------------------------------------------------------------------
# USER VE ITEM INDEX MAPPING
#
# CF algoritmalarında kullanıcı ve ürün ID'lerini doğrudan string/object
# olarak kullanmak yerine, bunları integer index'e çeviriyoruz.
#
# Amaç:
# - sparse matrix oluşturabilmek
# - her kullanıcıya ve ürüne bir satır/sütun numarası vermek
# ------------------------------------------------------------------

user_ids = cf_data["author_id"].unique()
item_ids = cf_data["product_id"].unique()

user_to_idx = {user_id: idx for idx, user_id in enumerate(user_ids)}
item_to_idx = {item_id: idx for idx, item_id in enumerate(item_ids)}

idx_to_user = {idx: user_id for user_id, idx in user_to_idx.items()}
idx_to_item = {idx: item_id for item_id, idx in item_to_idx.items()}

print("Toplam kullanıcı:", len(user_to_idx))
print("Toplam ürün:", len(item_to_idx))
```

```python
# ------------------------------------------------------------------
# USER VE ITEM INDEX KOLONLARI
#
# Her rating kaydını:
# - user_idx
# - item_idx
# şeklinde sayısal index'lere çeviriyoruz.
# ------------------------------------------------------------------

cf_data["user_idx"] = cf_data["author_id"].map(user_to_idx)
cf_data["item_idx"] = cf_data["product_id"].map(item_to_idx)

cf_data[["author_id", "product_id", "user_idx", "item_idx", "rating"]].head()
```

```python
# ------------------------------------------------------------------
# SPARSE USER-ITEM MATRIX
#
# Bu matriste:
# - satırlar kullanıcıları
# - sütunlar ürünleri
# - hücreler rating değerlerini temsil eder
#
# Veri çok seyrek olduğu için scipy'nin sparse matrix yapısını kullanıyoruz.
# ------------------------------------------------------------------

from scipy.sparse import csr_matrix

user_item_matrix = csr_matrix(
    (
        cf_data["rating"],
        (cf_data["user_idx"], cf_data["item_idx"])
    ),
    shape=(len(user_to_idx), len(item_to_idx))
)

print("User-item matrix shape:", user_item_matrix.shape)
print("Non-zero entries (rating sayısı):", user_item_matrix.nnz)
```

```python
# ------------------------------------------------------------------
# TRUNCATED SVD İLE LATENT FACTOR ÇIKARMA
#
# Amaç:
# User-item matrisini daha düşük boyutlu latent faktörlere indirgemek.
#
# Bu yaklaşım klasik collaborative filtering mantığına uygundur:
# - benzer kullanıcı davranışlarını öğrenir
# - benzer ürün yapılarını öğrenir
# - seyrek matriste çalışabilir
# ------------------------------------------------------------------

from sklearn.decomposition import TruncatedSVD

n_factors = 50  # başlangıç için makul latent factor sayısı

svd = TruncatedSVD(n_components=n_factors, random_state=42)
user_factors = svd.fit_transform(user_item_matrix)
item_factors = svd.components_.T

print("User latent matrix shape:", user_factors.shape)
print("Item latent matrix shape:", item_factors.shape)
```

```python
# ------------------------------------------------------------------
# TAHMİNİ USER-ITEM SKOR MATRİSİ
#
# Kullanıcı latent vektörleri ile ürün latent vektörlerini çarparak
# tahmini skor matrisi oluşturuyoruz.
#
# predicted_scores[user, item]
# → o kullanıcı için o ürünün tahmini skoru
# ------------------------------------------------------------------

predicted_scores = user_factors @ item_factors.T

print("Predicted score matrix shape:", predicted_scores.shape)
```

```python
# ------------------------------------------------------------------
# COLLABORATIVE FILTERING RECOMMENDATION FUNCTION
#
# Bu fonksiyon belirli bir kullanıcı için henüz puan vermediği ürünleri
# tahmini skora göre sıralar ve en iyi top_n ürünü önerir.
# ------------------------------------------------------------------

def recommend_cf_products(author_id, top_n=10):
    # Kullanıcı veri setinde yoksa bilgi ver
    if author_id not in user_to_idx:
        print("Bu kullanıcı CF veri setinde bulunamadı.")
        return None

    user_idx = user_to_idx[author_id]

    # Kullanıcının tahmini skorlarını al
    user_scores = predicted_scores[user_idx]

    # Kullanıcının daha önce puan verdiği ürünleri bul
    rated_items = set(
        cf_data.loc[cf_data["author_id"] == author_id, "product_id"].values
    )

    # Tüm ürünler için skor tablosu oluştur
    rec_df = pd.DataFrame({
        "product_id": item_ids,
        "cf_score": user_scores
    })

    # Daha önce puan verilen ürünleri çıkar
    rec_df = rec_df[~rec_df["product_id"].isin(rated_items)]

    # Ürün bilgileri ile birleştir
    rec_df = rec_df.merge(
        products_clean,
        on="product_id",
        how="left"
    )

    # En yüksek skorlu ürünleri sırala
    rec_df = rec_df.sort_values("cf_score", ascending=False)

    result_cols = [
        "product_id",
        "product_name",
        "brand_name",
        "primary_category",
        "secondary_category",
        "tertiary_category",
        "price_usd",
        "rating",
        "reviews",
        "loves_count",
        "cf_score"
    ]

    result_cols = [col for col in result_cols if col in rec_df.columns]

    return rec_df[result_cols].head(top_n)
```

```python
# ------------------------------------------------------------------
# UNION-BASED HYBRID RECOMMENDATION FUNCTION
#
# Bu fonksiyon content-based ve collaborative filtering önerilerini
# tek bir ortak listede birleştirir.
#
# Önceki hybrid yaklaşımında sadece content önerileri ana listeydi.
# Eğer CF önerileri ile content önerileri kesişmezse cf_score katkısı sıfır kalıyordu.
#
# Bu yeni yaklaşımda:
# - content önerileri alınır
# - CF önerileri alınır
# - iki liste union (birleşim) mantığıyla birleştirilir
# - eksik score alanları 0 ile doldurulur
# - iki skor normalize edilir
# - final hybrid score hesaplanır
# ------------------------------------------------------------------

def hybrid_recommendation(product_name, author_id, top_n=10, alpha=0.6):

    # --------------------------------------------------------------
    # 1) Content-based önerileri al
    # --------------------------------------------------------------
    content_rec = recommend_similar_products(product_name, top_n=50)

    # --------------------------------------------------------------
    # 2) Collaborative filtering önerileri al
    # --------------------------------------------------------------
    cf_rec = recommend_cf_products(author_id, top_n=50)

    if content_rec is None or cf_rec is None:
        print("Hybrid recommendation üretilemedi.")
        return None

    # --------------------------------------------------------------
    # 3) Content tarafında olmayan ama CF tarafında olan ürünleri de
    # korumak için outer merge kullanıyoruz
    # --------------------------------------------------------------
    hybrid_df = content_rec.merge(
        cf_rec[["product_id", "cf_score"]],
        on="product_id",
        how="outer"
    )

    # --------------------------------------------------------------
    # 4) Content tablosundan gelmeyen ürünlerde ürün bilgileri eksik olur
    # Bunları products_clean tablosundan tamamlıyoruz
    # --------------------------------------------------------------
    hybrid_df = hybrid_df.merge(
        products_clean[
            [
                "product_id",
                "product_name",
                "brand_name",
                "primary_category",
                "secondary_category",
                "tertiary_category",
                "price_usd",
                "rating",
                "reviews",
                "loves_count"
            ]
        ],
        on="product_id",
        how="left",
        suffixes=("", "_from_products")
    )

    # --------------------------------------------------------------
    # 5) Eğer bazı ürün bilgileri content_rec'den gelmediyse
    # products_clean'den gelen değerlerle dolduruyoruz
    # --------------------------------------------------------------
    for col in [
        "product_name",
        "brand_name",
        "primary_category",
        "secondary_category",
        "tertiary_category",
        "price_usd",
        "rating",
        "reviews",
        "loves_count"
    ]:
        fallback_col = f"{col}_from_products"
        if fallback_col in hybrid_df.columns:
            hybrid_df[col] = hybrid_df[col].fillna(hybrid_df[fallback_col])

    # Geçici kolonları temizle
    hybrid_df = hybrid_df.drop(
        columns=[col for col in hybrid_df.columns if col.endswith("_from_products")]
    )

    # --------------------------------------------------------------
    # 6) Eksik skorları 0 ile doldur
    # - content tarafından gelmeyen üründe similarity_score = 0
    # - CF tarafından gelmeyen üründe cf_score = 0
    # --------------------------------------------------------------
    hybrid_df["similarity_score"] = hybrid_df["similarity_score"].fillna(0)
    hybrid_df["cf_score"] = hybrid_df["cf_score"].fillna(0)

    # --------------------------------------------------------------
    # 7) Skorları normalize et
    #
    # similarity_score ve cf_score aynı ölçekte değildir.
    # Bu nedenle min-max normalization uyguluyoruz.
    # Eğer tüm değerler aynıysa, skorları 0 olarak bırakıyoruz.
    # --------------------------------------------------------------
    def min_max_normalize(series):
        min_val = series.min()
        max_val = series.max()

        if max_val == min_val:
            return pd.Series([0] * len(series), index=series.index)

        return (series - min_val) / (max_val - min_val)

    hybrid_df["content_score_norm"] = min_max_normalize(hybrid_df["similarity_score"])
    hybrid_df["cf_score_norm"] = min_max_normalize(hybrid_df["cf_score"])

    # --------------------------------------------------------------
    # 8) Final hybrid score hesapla
    #
    # alpha = content ağırlığı
    # (1 - alpha) = collaborative filtering ağırlığı
    # --------------------------------------------------------------
    hybrid_df["final_score"] = (
        alpha * hybrid_df["content_score_norm"] +
        (1 - alpha) * hybrid_df["cf_score_norm"]
    )

    # --------------------------------------------------------------
    # 9) Sonuçları sırala
    # --------------------------------------------------------------
    hybrid_df = hybrid_df.sort_values("final_score", ascending=False)

    # --------------------------------------------------------------
    # 10) Sonuç tablosunda gösterilecek sütunlar
    # --------------------------------------------------------------
    result_cols = [
        "product_id",
        "product_name",
        "brand_name",
        "primary_category",
        "secondary_category",
        "tertiary_category",
        "price_usd",
        "rating",
        "reviews",
        "loves_count",
        "similarity_score",
        "cf_score",
        "content_score_norm",
        "cf_score_norm",
        "final_score"
    ]

    result_cols = [col for col in result_cols if col in hybrid_df.columns]

    return hybrid_df[result_cols].head(top_n)
```
